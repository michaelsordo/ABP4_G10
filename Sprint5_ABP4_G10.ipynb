# **Preprocesado y carga de datasets**


El dataset con el que trabajamos cuenta con una lista de 5000 juegos, todos ellos con un campo 'short_description' que contiene una descripción corta de cada uno. En las siguientes celdas de código se transforman los textos contenidos en el campo mencionado anteriormente, a texto procesado. Estes nuevos textos tokenizados se almacenan en una nueva columna a la que llamamos 'short_description_processed'.





```python

import pandas as pd

originalData = pd.read_csv('steam.csv')
originalData = originalData.head(1000)
originalData
```

```python
from nltk.tokenize import word_tokenize 
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk
nltk.download('punkt')
nltk.download('stopwords')

ps = PorterStemmer()

preprocessedText = []

for row in originalData.itertuples():
    
    text = word_tokenize(row[17]) ## indice de la columna que contiene el texto
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [ps.stem(w) for w in text if not w in stops and w.isalnum()]
    text = " ".join(text)
    
    preprocessedText.append(text)

preprocessedData = originalData
preprocessedData['short_description_processed'] = preprocessedText

preprocessedData
```

# **Creación de la bolsa de palabras (BoW) con TF-IDF**

Se parte de los datos almacenados en "preprocessedData", en donde para cada videojuego existe un campo 'shor_description_processed' que contiene la descripción corta preprocesada.

El objetivo es transformar todos los textos de descripciones cortas en vectores de frecuencias (Bag of words), aplicando además la ponderación TF-IDF para los valores de dichas frecuencias.

El paquete sklearn ofrece una clase llamada TfidfVectorizer que crea automáticamente la matriz compuesta por todos los vectores de frecuencias ponderados a partir de un array de textos (preprocessedData['short_description_processed'])

Si se quiere emplear la bolsa de palabras sin ponderación TF-IDF puede usarse la clase CountVectorized del mismo paquete



```python
from sklearn.feature_extraction.text import TfidfVectorizer

bagOfWordsModel = TfidfVectorizer()
bagOfWordsModel.fit(preprocessedData['short_description_processed'])
textsBoW= bagOfWordsModel.transform(preprocessedData['short_description_processed'])
```

# **Cálculo de distancias entre vectores de frecuencias**

El objetivo final es el de crear una matriz N x N (N= número de videojuegos) en donde el valor de la posición matriz[i,j] indique la distancia que existe entre el videojuego i y el videojuego j.

Esta distancia se puede calcular de varias formas. Gracias a que ahora los textos están representados mediante vectores de frecuencias (textsBoW), se pueden emplear para ello medidas de distancias standard entre vectores. La que usamos es la distancia coseno.

```python
from sklearn.metrics import pairwise_distances

distance_matrix= pairwise_distances(textsBoW,textsBoW ,metric='cosine')
```

# **Búsqueda de los videojuegos más similares a otros en base a la descripción corta**

```python
searchTitle = "Dangerous Waters" #Juego base para la recomendación
indexOfTitle = preprocessedData[preprocessedData['name']==searchTitle].index.values[0]
indexOfTitle
```

```python
distance_scores = list(enumerate(distance_matrix[indexOfTitle]))
distance_scores
```

```python
ordered_scores = sorted(distance_scores, key=lambda x: x[1])
ordered_scores
```

#**Algoritmo de clasificación supervisado usando un dataset para el análisis de sentimientos**

El análisis de sentimientos nos permitirá analizar los comentarios sobre los videojuegos, y en función de estos, sean positivos o negativos, hacer su posible recomendación. Contamos para esto con un dataset de **10000** filas.

```python
trainingData = pd.read_csv('steam_reviews_train.csv', delimiter=',')
trainingData
```

```python
trainingData['recommendation'].value_counts()
```

```python
trainingData['review'] = trainingData['review'].apply(str)
```

# **Preprocesamiento de los datos de entrenamiento**

En las siguientes celdas de código se transforman los textos contenidos a texto procesado. 



```python
ps = PorterStemmer()

preprocessedText = []

for row in trainingData.itertuples():
    
    
    text = word_tokenize(row[8]) ## indice de la columna que contiene el texto
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [ps.stem(w) for w in text if not w in stops and w.isalnum()]
    text = " ".join(text)
    
    preprocessedText.append(text)

preprocessedData = trainingData
preprocessedData['review_processed'] = preprocessedText

preprocessedData
```

#**Creación de la bolsa de palabras**

Partimos de los datos almacenados en "preprocessedData".

El paquete sklearn ofrece una clase llamada TfidfVectorizer que crea automáticamente la matriz compuesta por todos los vectores de frecuencias ponderados a partir de un array de textos.

```python
bagOfWordsModel = TfidfVectorizer()
bagOfWordsModel.fit(preprocessedData['review_processed'])
textsBoW= bagOfWordsModel.transform(preprocessedData['review_processed'])
print("Finished")
```

```python
textsBoW.shape
```

#**Entrenamiento de un algoritmo de clasificación (SVM)**

SVM es un modelo que representa a los puntos de muestra en el espacio, separando las clases a dos espacios lo más amplios posibles mediante un hiperplano de separación definido como el vector entre los dos puntos, de las dos clases, más cercanos al que se llama vector soporte. 

Nosotros generaremos un modelo basado en SVM con el tipo de kernel lineal.

```python
from sklearn import svm
svc = svm.SVC(kernel='linear') #Modelo de clasificación

X_train = textsBoW #Documentos
Y_train = trainingData['recommendation'] #Etiquetas de los documentos 
svc.fit(X_train, Y_train) #Entrenamiento
```

#**Carga de y preprocesado de documentos de test**

```python
testData = pd.read_csv('steam_reviews_test.csv', delimiter=',')

testData
```

```python
testData['review'] = testData['review'].apply(str)
```

```python
ps = PorterStemmer()

preprocessedText = []

for row in testData.itertuples():
    
    
    text = word_tokenize(row[8]) ## indice de la columna que contiene el texto
    ## Remove stop words
    stops = set(stopwords.words("english"))
    text = [ps.stem(w) for w in text if not w in stops and w.isalnum()]
    text = " ".join(text)
    
    preprocessedText.append(text)

preprocessedDataTest = testData
preprocessedDataTest['processed_text'] = preprocessedText

preprocessedDataTest
```

```python
testData['recommendation'].value_counts()
```

```python
textsBoWTest= bagOfWordsModel.transform(preprocessedDataTest['processed_text'])
print("Finished")
```

```python
textsBoWTest.shape
```

#**Clasificación de los documentos de test**

Almacenamos en el array predictions las predicciones del clasificador.

```python
X_test = textsBoWTest #Documentos


predictions = svc.predict(X_test) 
```

#**Evaluación de la predicción**

```python
from sklearn.metrics import classification_report

Y_test = testData['recommendation'] #Etiquetas reales de los documentos

print (classification_report(Y_test, predictions))
```

#**Entrenamiento y Evaluación de otro algoritmo de clasificación: k-NN**

El clasificador k-vecinos más próximos, en inglés k-nearest neightbours, se basa en la suposición de que se puede clasificar un elemento en función a los k elementos más cercanos que tenga, siendo k un número entero no predefinido.

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)


neigh.fit(X_train, Y_train) 
predictions = neigh.predict(X_test) 

print (classification_report(Y_test, predictions))
```
